# Removendo todos os registros onde as informações do Bairro são nulas

```python
sem_bairro = base.dropna(subset=['bairro'])
sem_bairro
```

## Filtrando hotéis que se adequam ao número de noites de estadia

```python
#excluindo do dataframe os hoteis que exigem um número de noites mínimas maior que 4, 
base_NoitesMinimas = sem_bairro[sem_bairro.noites_mínimas <= 4]

#e também excluindo os hoteis que não permitem estadias menores que 4
base_NoitesAdequadas = base_NoitesMinimas[base_NoitesMinimas.noites_máximas >= 4]

#printando base atualizada
base_NoitesAdequadas #base_NoitesAdequadas consiste no DataFrame com o número de noites adequadas
```

# Filtrar Fumetas
Filtrando os registros, onde a coluna "Facilidades" possui o item "smoking Allowed"

```python
#filtrando o dataframe anterior(dataframe com o número de noites adequadas), e procurando pelos registros onde é permitido fumar
baseFumeta = base_NoitesAdequadas[base_NoitesAdequadas['facilidades'].str.contains("Smoking allowed")] 
baseFumeta
```

# Filtrando pelos hotéis que possuem Internet, Wifi ou Wi-Fi, visando a necessidade de 1 dia de trabalho do hóspede

Ninguém merece trabalhar no computador da Lan House

```python
# procurando por 3 termos dentro da coluna facilidades, através de diversos operadores OR
baseInternet = baseFumeta[baseFumeta['facilidades'].str.contains("Internet") | baseFumeta['facilidades'].str.contains("Wifi") | baseFumeta['facilidades'].str.contains("Wi-Fi")] 
baseInternet
```
